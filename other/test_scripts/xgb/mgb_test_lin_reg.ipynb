# MGBoost Linear Regression
Testing MGBoost with a multiobjective linear regression dataset

```python
import numpy as np
import pandas as pd
from xgboost import XGBRegressor
from xgboost.sklearn import fobj
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
from matplotlib import style; style.use('ggplot')
```

```python
data = pd.read_csv('/Users/Ashtekar15/Desktop/Thesis/MGBoost/other/test_data/linear_regression/linear_regression_data.csv')
X = data.X.values.reshape(-1, 1)
y = data.y.values
```

```python
xgbr = XGBRegressor(objective=fobj)
xgbr.fit(X, y)
y_pred = xgbr.predict(X)
```

```python
plt.rcParams['figure.figsize'] = [10, 6]

plt.scatter(X, y, color='green', alpha=0.5, label='y')
# plt.scatter(X, 5 * y, color='blue', alpha=0.5, label='5 * y')
plt.scatter(X, 10 * y, color='red', alpha=0.5, label='10 * y')
plt.scatter(X, y_pred, color='black', alpha=1, label='y_pred');
plt.legend()
plt.title('Multiobjective Linear Regression');
```

```python
plt.rcParams['figure.figsize'] = [10, 6]

plt.scatter(X, 5 * y, color='blue', alpha=0.5, label='5 * y')
plt.scatter(X, y_pred, color='black', alpha=1, label='y_pred');
plt.legend()
plt.title('Predictions and 5 * y Plotted');
```

MGBoost learns to approximate the function `X, 5 * y` when given objectives `y` and `10 * y`.

This is what we would expect: `5 * y` is a pareto-optimal solution given the objectives.
